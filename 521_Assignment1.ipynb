```python
# Use OLS regression to make 3 specifications using the data from df

import statsmodels.formula.api as sm
#create tinc squared (for square of total incentive):
df['tinc_sq'] = df['tinc']**2

# Specification 1
result1 = sm.ols('got ~ any + hiv2004 + male + age + rumphi + balaka', data=df).fit()
print(result1.summary())

# Specification 2
result2 = sm.ols('got ~ any + tinc + hiv2004 + male + age + rumphi + balaka', data=df).fit()
print(result1.summary())


# Specification 3
result3 = sm.ols('got ~ any + tinc + tinc_sq + hiv2004 + male + age + rumphi + balaka', data=df).fit()
print(result1.summary())
```

```python
# Show the three results in a stargazer table

!pip install stargazer
from stargazer.stargazer import Stargazer

stargazer = Stargazer([result1, result2, result3])

stargazer.covariate_order(['any', 'tinc', 'tinc_sq', 'hiv2004', 'male', 'age','rumphi', 'balaka', 'Intercept'])

 #Rename the covariates
stargazer.rename_covariates({
    'any': 'Any incentive',
    'tinc': 'Amount of incentive',
    'tinc_sq': 'Amount of incentive sq',
    'hiv2004': 'HIV',
    'male': 'Male',
    'age': 'Age',
    'rumphi': 'Rumphi',
    'balaka': 'Balaka',
    'Intercept': 'Constant'

})

# Rename the dependent variable
stargazer.dependent_variable_name('Attendance at HIV results centers')

stargazer.title("TABLE 4- IMPACT OF MONETARY INCENTIVES AND DISTANCE ON LEARNING HIV TEST RESULTS")

stargazer
```

```python
# prompt: render table as latex

print(stargazer.render_latex())
```

```python
# prompt: make a variable called yearsexp10 and yearsexp100 using yearsexp

df1['yearsexp10'] = df1['yearsexp'] * 10
df1['yearsexp100'] = df1['yearsexp10'] ** 2
```

```python
# prompt: divide the df1 by race

# Divide the df1 by race
white_df = df1[df1['race'] == 0]
black_df = df1[df1['race'] == 1]
```

```python
# prompt: create years of experience squared

#create years of experience squared:
df1['yearsexp_sq'] = df1['yearsexp']**2
white_df['yearsexp_sq'] = white_df['yearsexp']**2
black_df['yearsexp_sq'] = black_df['yearsexp']**2
```

```python
#Use OLS to replicate the first 3 columns of Table 5 in the paper:

# Specification 1
result4 = sm.ols('call ~ yearsexp + yearsexp_sq + volunteer + military + email + empholes + workinschool + honors + computerskills + specialskills', data=df1).fit()
print(result1.summary())

# Specification 2
result5 = sm.ols('call ~ yearsexp + yearsexp_sq + volunteer + military + email + empholes + workinschool + honors + computerskills + specialskills', data=white_df).fit()
print(result1.summary())

# Specification 3
result6 = sm.ols('call ~ yearsexp + yearsexp_sq + volunteer + military + email + empholes + workinschool + honors + computerskills + specialskills', data=black_df).fit()
print(result1.summary())
```

```python
# Create a function to modify the coefficients
def modify_coefficients(result):
    params = result.params.copy()
    params['yearsexp'] = params['yearsexp'] * 10
    params['yearsexp_sq'] = params['yearsexp_sq'] * 100
    return params
```

```python
# Apply the function to each result
result4.params = modify_coefficients(result4)
result5.params = modify_coefficients(result5)
result6.params = modify_coefficients(result6)
```

```python
stargazer = Stargazer([result4, result5, result6])

stargazer.covariate_order(['yearsexp', 'yearsexp_sq', 'volunteer', 'military', 'email','empholes', 'workinschool', 'honors', 'computerskills', 'specialskills'])
```

```python
# Rename the covariates
stargazer.rename_covariates({
    'yearsexp': 'Years of Experience (*10)',
    'yearsexp_sq': 'Years of Experience Squared (*100)',
    'volunteer': 'Volunteering? (Y=1)',
    'military': 'Military Experience? (Y=1)',
    'email': 'Email? (Y=1)',
    'empholes': 'Employment Gaps? (Y=1)',
    'workinschool': 'Work in School? (Y=1)',
    'honors': 'Honors? (Y=1)',
    'computerskills': 'Computer Skills? (Y=1)',
    'specialskills': 'Special Skills? (Y=1)'
})

# Rename the specifications
stargazer.custom_columns(['All Resumes', 'White Names', 'African-American Names'], [1, 1, 1])

# Rename the dependent variable
stargazer.dependent_variable_name('Callback Dummy')

stargazer.title("TABLE 5- EFFECT OF RESUME CHARACTERISTICS ON LIKELIHOOD OF CALLBACK")
```
